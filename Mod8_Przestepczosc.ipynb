Ad 1. Pobierz bazę danych dotyczącą śmiertelnych interwencji policji w USA, a następniewczytaj ją do obiektu DataFrame.

```python
import pandas as pd
df = pd.read_csv('fatal-police-shootings-data.csv')
df.head()
```

Ad 2. Przekształć tabelę w taki sposób, aby wskazywała zestawienie jednocześnie liczby ofiar interwencji według rasy (‘race’) oraz tego, czy wykazywały one oznaki chorobypsychicznej (‘signs_of_mental_illness’).

```python
import numpy as np
df_t=df.pivot_table(values='name',index='race',columns='signs_of_mental_illness',aggfunc=len)
df_t
```

Ad 3. Za pomocą Map, Applymap lub Apply dodaj do tego zestawienia kolumnę wskazującą jaki odsetek ofiar interwencji wykazywało oznaki choroby psychicznej dla każdej z ras. Odpowiedz, która z nich charakteryzuje się największym odsetkiem znamion choroby psychicznej podczas interwencji.

```python
df_t.columns =['False', 'True']
```

```python
def percent_mental(row):
     return row['True']/(row['True']+row['False'])
```

```python
df_t['Mental Percent'] = df_t.apply(lambda row: percent_mental(row),axis=1).round(2)
```

Ad 4. Dodaj kolumnę oznaczającą dzień tygodnia, w którym doszło do interwencji. Zlicz interwencje według odpowiedniego dnia tygodnia. Następnie stwórz wykres kolumnowy, tak aby dni tygodnia były uszeregowane od poniedziałku do niedzieli.

```python
df['date'] = pd.to_datetime(df['date'])
df['day']=df['date'].dt.day_name()
df['day'].value_counts()
```

```python
df['day']=pd.Categorical(df['day'], categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'], ordered=True)
```

```python
df_p=df['day'].value_counts().sort_index()
plot1 = df_p.plot(kind='bar',legend=None,title="Ilość interwencji w dniu")
plot1.set_xlabel("Dzień tygodnia")
plot1.set_ylabel("Interwencje")
```

Ad 5. Wczytaj do projektu dane dotyczące populacji w poszczególnych stanach USA oraz dane dotyczące skrótów poszczególnych stanów. Połącz te bazy danych w taki sposób, aby móc policzyć do ilu incydentów w bazie dotyczącej śmiertelnych interwencji doszło w przeliczeniu na 1000 mieszkańców każdego ze stanów.

```python
dat1=pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population', header=0)
dat2=pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations', header=0)
```

```python
populacja=dat1[0]
kody=dat2[0]
```

```python
kody.rename(columns={ kody.columns[0]: 'state_name', kody.columns[5]: 'state'}, inplace = True)
kody=kody[['state_name', 'state']]
```

```python
populacja.rename(columns={ populacja.columns[2]: 'state_name', populacja.columns[3]: 'populacja'}, inplace = True)
populacja=populacja[['state_name', 'populacja']]
```

```python
right=pd.merge(populacja,kody,how='inner',on='state_name')
```

```python
df=pd.merge(df,right,how='inner',on='state')
```

```python
tab1=df.pivot_table(values='name',index=['state_name', 'populacja'], aggfunc=len).reset_index()
```

```python
tab1.rename(columns={tab1.columns[2]: 'interventions'}, inplace = True)
tab1
```

```python
def interv1(row):
     return row['interventions']/(0.001*row['populacja'])
```

```python
tab1['Interventions per 1000'] = tab1.apply(lambda row: interv1(row),axis=1).round(2)
```
